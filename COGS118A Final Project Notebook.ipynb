```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.preprocessing import OrdinalEncoder
```

```python
credit_approval = X_credit_approval.assign(**{"A16":y_credit_approval["A16"]})
```

```python
credit_approval = credit_approval.dropna()
```

```python
credit_approval = credit_approval.assign(**{"A16":credit_approval["A16"].replace({"+": 1, "-": 0})})
credit_approval
```

```python
# List of categorical columns
object_cols = ['A13','A12','A10','A9','A7','A6','A5','A4','A1']

# ordinal-encode categorical columns
X = credit_approval.copy()
ordinal_encoder = OrdinalEncoder()
X[object_cols] = ordinal_encoder.fit_transform(credit_approval[object_cols])

# Preview the ordinal-encoded features
X.head()
```

```python
def experiment(data, dataset_name):

    # Split the data into three partitions (20/80, 50/50, 80/20)
    partitions = [(0.2, 0.8), (0.5, 0.5), (0.8, 0.2)]

    # Define classifiers
    classifiers = [
        ('Logistic Regression', LogisticRegression()),
        ('Decision Tree', DecisionTreeClassifier()),
        ('Random Forest', RandomForestClassifier()),
    ]
    
    if dataset_name == 'credit_approval':
        X = data[["A15", "A14", "A9", "A3"]]
        y = data["A16"]
    elif dataset_name == 'adult':
        X = data[["age", "education-num", "marital-status", "sex", "hours-per-week", "capital-gain"]]
        y = data["income"]
    elif dataset_name == "heart_disease":
        X = data[["cp", "ca", "thal", "exang", "thalach", "oldpeak"]]
        y = data["num"]
                      
    classifier_performance = []
    # Perform the iterations
    for classifier_name, classifier in classifiers:
        print(f"Classifier: {classifier_name}")
        for train_ratio, test_ratio in partitions:
            # Split the data into train, validation, and test sets
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_ratio)
            X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=1/3)

            # Assuming you're tuning hyperparameters here, modify this part based on your classifier
            # Example for Decision Tree Classifier hyperparameter tuning
            if classifier_name == 'Logistic Regression':
                
                param_grid = {'C': [0.001, 0.01, 0.1, 1, 10], 'penalty': ['l1', 'l2'], 
                              'solver': ['liblinear', 'lbfgs', 'newton-cg', 'sag', 'saga']}
                
                # Instantiate the logistic regression model
                logistic_regression = LogisticRegression(class_weight='balanced')

                # Perform grid search with cross-validation
                grid_search = GridSearchCV(logistic_regression, param_grid, cv=3, scoring='accuracy')
                grid_search.fit(X_train, y_train)

                # Get the best parameters and best score
                best_params = grid_search.best_params_
                best_score = grid_search.best_score_

                # Train the model using the best hyperparameters found
                final_clf = LogisticRegression(**best_params)
                final_clf.fit(X_train, y_train)

                # Evaluate performance on train, validation, and test sets
                train_acc = final_clf.score(X_train, y_train)
                val_acc = final_clf.score(X_val, y_val)
                test_acc = final_clf.score(X_test, y_test)
                
                split = {"Classifier": classifier_name, "Partition": f"{train_ratio*100}/{test_ratio*100}", 
                                     "Best Parameter": best_params, "Train Acc": train_acc, 
                                     "Val Acc": val_acc, "Test Acc": test_acc}
                
                classifier_performance.append(split)

                print(f"Partition: {train_ratio*100}/{test_ratio*100} - Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}, Test Acc: {test_acc:.4f}")
            
            elif classifier_name == 'Decision Tree':

                param_grid = {
                'max_depth': [None, 5, 10, 15],
                'min_samples_split': [2, 5, 10],
                'min_samples_leaf': [1, 2, 4],
                'criterion': ['gini', 'entropy']
                }

                # Instantiate the Decision Tree classifier
                decision_tree = DecisionTreeClassifier(class_weight='balanced')

                # Perform grid search with cross-validation
                grid_search = GridSearchCV(decision_tree, param_grid, cv=3, scoring='accuracy')
                grid_search.fit(X_train, y_train)

                # Get the best parameters and best score
                best_params = grid_search.best_params_
                best_score = grid_search.best_score_

                # Train the model using the best hyperparameters found
                final_clf = DecisionTreeClassifier(**best_params)
                final_clf.fit(X_train, y_train)

                # Evaluate performance on train, validation, and test sets
                train_acc = final_clf.score(X_train, y_train)
                val_acc = final_clf.score(X_val, y_val)
                test_acc = final_clf.score(X_test, y_test)
                
                split = {"Classifier": classifier_name, "Partition": f"{train_ratio*100}/{test_ratio*100}", 
                                     "Best Parameter": best_params, "Train Acc": train_acc, 
                                     "Val Acc": val_acc, "Test Acc": test_acc}
                
                classifier_performance.append(split)

                print(f"Partition: {train_ratio*100}/{test_ratio*100} - Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}, Test Acc: {test_acc:.4f}")

            else:

                param_grid = {
                'n_estimators': [50, 100, 200],
                'max_depth': [None, 5, 10, 15],
                'min_samples_split': [2, 5, 10],
                'min_samples_leaf': [1, 2, 4],
                'criterion': ['gini', 'entropy']
                }

                # Instantiate the Random Forest classifier
                random_forest = RandomForestClassifier(class_weight='balanced')

                # Perform grid search with cross-validation
                grid_search = GridSearchCV(random_forest, param_grid, cv=3, scoring='accuracy')
                grid_search.fit(X_train, y_train)

                # Get the best parameters and best score
                best_params = grid_search.best_params_
                best_score = grid_search.best_score_


                # Train the model using the best hyperparameters found
                final_clf = RandomForestClassifier(**best_params)
                final_clf.fit(X_train, y_train)

                # Evaluate performance on train, validation, and test sets
                train_acc = final_clf.score(X_train, y_train)
                val_acc = final_clf.score(X_val, y_val)
                test_acc = final_clf.score(X_test, y_test)
                
                split = {"Classifier": classifier_name, "Partition": f"{train_ratio*100}/{test_ratio*100}", 
                                     "Best Parameter": best_params, "Train Acc": train_acc, 
                                     "Val Acc": val_acc, "Test Acc": test_acc}
                
                classifier_performance.append(split)

                print(f"Partition: {train_ratio*100}/{test_ratio*100} - Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}, Test Acc: {test_acc:.4f}")
            
    return pd.DataFrame(classifier_performance)
```
